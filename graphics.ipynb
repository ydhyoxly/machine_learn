Считать реальное изображение можно при помощи функции scipy.misc.imread (если изображение не в формате png, установите пакет pillow).


```python
%matplotlib inline
import matplotlib.pyplot as plt
import time
import numpy as np
import pandas as p

data_size = np.array([30, 56, 104, 210, 320, 548, 980, 1670])
time_non_vectorized = np.zeros(8)
time_vectorized = np.zeros(8)

for s in range(data_size.size):
    
    img_arr = np.random.randint(0, 256, (data_size[s], data_size[s], 3), dtype=np.uint8)
    gray_weights = np.array([0.299, 0.587, 0.114])
    
    start_1 = time.time() 

    def vector_summ(image, weights):
    
        return np.tensordot(image, weights, axes=([2], [0]))
    
    
    gray_img_2 = vector_summ(img_arr, gray_weights)
    
    end_1 = time.time() - start_1
    time_vectorized[s] = end_1*1000000
    
    img_arr_1 = img_arr.tolist()
    gray_weights_1 = gray_weights.tolist()
    
    start_1 = time.time() 

    def summ(image, weights):
        result = [[0 for i in range(len(image[0]))] for j in range(len(image))]
        
        
        for i in range(len(image)):
            for j in range(len(image[0])):
                for e in range(len(image[0][0])):
                    
                    result[i][j] += (weights[e]*image[i][j][e])
                result[i][j] = round(result[i][j], 1)
            
    
        return result
    
    
    gray_img_1 = summ(img_arr_1, gray_weights_1)
    
    end_1 = time.time() - start_1 
    time_non_vectorized[s] = end_1*1000000

    
    
fig = plt.figure()

ax2 = fig.add_subplot(111)


ax2.plot(data_size, time_non_vectorized, label=u"Невекторизованная")
ax2.plot(data_size, time_vectorized, label=u"Векторизованная")
ax2.set_title(u"4 задача")
ax2.set_xlabel(u"Длина массива")
ax2.set_ylabel(u"Время (мкс)")
ax2.grid()
ax2.legend()

plt.show()
plt.close()
```

* __Задача 5__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ – (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
%matplotlib inline
import matplotlib.pyplot as plt
import time
import numpy as np
import pandas as p

data_size = np.array([200, 500, 1000, 2300, 3000, 5890, 7560, 10000])
time_non_vectorized = np.zeros(8)
time_vectorized = np.zeros(8)

for s in range(data_size.size):
    
    x = np.random.randint(0, 30, data_size[s])
    
    start_1 = time.time() 

    x1, x2 = np.unique(x, return_counts=True)
    
    end_1 = time.time() - start_1
    time_vectorized[s] = end_1*1000000
    
    x = x.tolist()
    
    start_1 = time.time() 

    dict = {}
    x1, x2 =[], []
    for i in range(len(x)):
        if x[i] not in dict:
            dict[x[i]]=1
        else:
            dict[x[i]]+=1
    for i in dict.keys():
        x1.append(i)
        x2.append(dict[i])
    
    end_1 = time.time() - start_1 
    time_non_vectorized[s] = end_1*1000000

    
    
fig = plt.figure()

ax2 = fig.add_subplot(111)


ax2.plot(data_size, time_non_vectorized, label=u"Невекторизованная")
ax2.plot(data_size, time_vectorized, label=u"Векторизованная")
ax2.set_title(u"5 задача")
ax2.set_xlabel(u"Длина массива")
ax2.set_ylabel(u"Время (мкс)")
ax2.grid()
ax2.legend()

plt.show()
plt.close()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import time
import numpy as np
from math import sqrt
import pandas as p

data_size = np.array([30, 56, 104, 210, 320, 548, 980, 1670])
time_non_vectorized = np.zeros(8)
time_vectorized = np.zeros(8)

for s in range(data_size.size):
    
    X = np.random.rand(data_size[s], 3)
    Y = np.random.rand(data_size[s], 3)
    
    start_1 = time.time()

    def vector_evc_dist(X, Y):
   
        X_2 = np.sum(X ** 2, axis=1, keepdims=True)
        
        Y_2 = np.sum(Y ** 2, axis=1)
        
        XY = np.dot(X, Y.T)
        return np.sqrt(X_2 - 2 * XY + Y_2)
    
    result = vector_evc_dist(X, Y)
    
    end_1 = time.time() - start_1
    time_vectorized[s] = end_1*1000000
    
    X_1 = X.tolist()
    Y_1 = Y.tolist()
    
    start_1 = time.time() 

    def evc_dist(X, Y):
    
        distances = [[0 for i in range(len(Y))] for j in range(len(X))]
        for i in range(len(X)):
            for j in range(len(Y)):
                for e in range(len(X[0])):
                    
                     distances[i][j] += (X[i][e] - Y[i][e])**2
                distances[i][j] = round(sqrt(distances[i][j]), 8)
        return distances
    
    evc_dist(X_1, Y_1)
    
    end_1 = time.time() - start_1 
    time_non_vectorized[s] = end_1*1000000

    
    
fig = plt.figure()

ax2 = fig.add_subplot(111)


ax2.plot(data_size, time_non_vectorized, label=u"Невекторизованная")
ax2.plot(data_size, time_vectorized, label=u"Векторизованная")
ax2.set_title(u"6 задача")
ax2.set_xlabel(u"Длина массива")
ax2.set_ylabel(u"Время (мкс)")
ax2.grid()
ax2.legend()

plt.show()
plt.close()
```
